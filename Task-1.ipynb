## Data Visualization

### Plotting Histogram of the data from 2000 - 2022

```python
years = df.columns[41:]
```

```python
import matplotlib.pyplot as plt
import numpy as np
years = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008',
         '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017',
         '2018', '2019', '2020', '2021', '2022']


num_rows = len(years) // 3 + (len(years) % 3 > 0)
num_cols = 3
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
axes = axes.flatten()
for i, year in enumerate(years):
    if i < len(axes):
        axes[i].hist(df[year], color='cyan', bins=10)
        axes[i].set_xlabel(year)
        axes[i].set_ylabel('FREQUENCY')
        axes[i].set_title(f'Histogram for {year}')

plt.tight_layout()
plt.show()
```

## Plotting Barplot For Each Year

```python
#Sum the values for each year across all countries
total_values = df[years].sum()

#Plotting the bar chart
plt.figure(figsize = (20,10))
plt.bar(years,total_values,color = 'blue')
plt.xlabel('Total Values')
plt.ylabel('Year', size = 20)
plt.title('Total values per Year', size = 20)
plt.show()
```

## Plotting Indicators of top 10 Countries for different years with Base year 1960

```python
country_1960 = df.sort_values(by = '1960').head(10)
country_1960.info()
```

⚫ NOTE:- There is one column (1970) which has some cells as object type because it contains some string and blank values. so remove that rows from the data set and then go ahead. I had removed it directly from excel using filter method.

```python

import matplotlib.pyplot as plt
import seaborn as sns
country_1960_t = country_1960.set_index('Country Name').T

num_rows = 21
num_cols = 3
total_height = num_rows * 5
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, total_height))

axes = axes.flatten()

for i, (year, data_values) in enumerate(country_1960_t.iterrows()):
    ax = axes[i]
    sns.barplot(x=data_values.index, y=data_values.values, ax=ax)
    ax.set_xlabel('Countries')
    ax.set_ylabel('Data Values')
    ax.set_title(f"{year} - Data Values From 1960 to 2022")
    ax.tick_params(axis='x', rotation=45)
    ax.set_xticklabels(ax.get_xticklabels(), ha='right')  

plt.tight_layout()
plt.show()
```
